```python
class RegularizedLogisticRegressionModel:
    def __init__(self, lr=0.01, num_iter=100000, fit_intercept=True,lagrange_mul=1):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.lagrange_mul=lagrange_mul
    
    def add_intercept(self, X):
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    def sigmoid(self, z):        
        return 1 / (1 + np.exp(-z))
        
    def grad(self, X, y,w,w0,lagrange_mul):
        N=X.shape[0]
        #print(w0)
        tmp=0
        tmp2=0
        
        for i in range(N):
            z=X[i].dot(w)+w0
            f_x_i=self.sigmoid(z)
            tmp +=f_x_i-y[i]
            #print(tmp)
            tmp3=np.array([X[i]])
            #print(tmp3.shape)
            tmp2 +=(tmp3.T).dot((f_x_i-y[i]))
        
        tmp2=np.array([tmp2])
        tmp2=tmp2.T
        
        w0=w0-(self.lr/N)*(tmp)
        w=w*(1-(self.lr*lagrange_mul)/N) - (self.lr/N)*(tmp2)
        return w,w0 
    
    def predict_prob(self, X,w):
        if self.fit_intercept:
            X = self.add_intercept(X)
        #w=w/X.shape[0]
        z=np.dot(X, w)
        
        result=self.sigmoid(z)
        #print(result)
        return result
    
    def predict(self, X, threshold,w):
        return self.predict_prob(X,w) >= threshold
    
    def loss(self, f_x, y,w,lagrange_mul):
        n=w.shape[0]
        N=f_x.shape[0]
        
        tmp=0
        tmp2=0
        for i in range(1,n):
            tmp +=w[i]**2
        for i in range(N):
            tmp2 +=y[i]*np.log(f_x[i]) + (1-y[i])*np.log(1-f_x[i])
        j_w= -tmp2+lagrange_mul*tmp       
        return j_w/N
    
    def write_prediction(self,filename,preds):
        param_file=open(filename,"w")
        for i in range(0,len(preds)):
            if(preds[i]==True):
                param_file.write(str(1))
            else:
                param_file.write(str(0))
            param_file.write("\n")
        param_file.close()
    
    def get_classifications(self,y,preds):
        preds_val=[]
        tp=0
        tn=0
        fp=0
        fn=0
        for i in range(len(preds)):
            if(preds[i] == True):
                preds_val.append(1)
            else:
                preds_val.append(0)
            if(preds_val[i] ==y[i]):
                if(preds_val[i]==1):
                    tp=tp+1
                else:
                    tn=tn+1
            else:
                if(preds_val[i]==1):
                    fp=fp+1
                else:
                    fn=fn+1  
        
        return tp,tn,fp,fn   
    def get_perf_params(self,tp,tn,fp,fn):
        sensitivity=0
        specificity=0
        precision=0
        f_measure=0
        if(tp+fn !=0):
            sensitivity=tp/(tp+fn)
        if(tn+fp !=0):
            specificity=tn/(tn+fp)
        if(tp+fp !=0):
            precision=tp/(tp+fp)
        if(precision + sensitivity !=0):
            f_measure=(2*precision*sensitivity)/(precision+sensitivity)
        accuracy=(tp+tn)/(tp+tn+fp+fn)
        
        return accuracy,sensitivity,specificity,precision,f_measure
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib notebook 
#to enable interactivity and rotation of 3D figures 
from mpl_toolkits.mplot3d import Axes3D
import math
from scipy.spatial import distance

class KNN:
    def __init__(self):
        pass
    def seperate_classes(self,X,y):
        tmp=y.tolist()
        flattened = [val for sublist in tmp for val in sublist]
        class_names = list(set(flattened))
        k=len(class_names)
        print("No of classes:=",k)
        class_list=[]
        for i in range(len(class_names)):
            tmplist=[]
            class_list.append(tmplist)
        print(class_names)
        for i in range(y.size):
            for j in range(len(class_names)):
                if(y[i] == class_names[j]):
                    class_list[j].append(X[i])
                    break
                
        for i in range(len(class_names)):
            class_list[i]=np.asarray(class_list[i])
           
        return class_names,class_list
    
    def find_co_variance(self,X,y,mean_list,class_names):
        class_list=[]
        k=len(class_names)
        n=X.shape[1]
        t=np.zeros((k,n,1))
        t_mean=np.zeros((k,n,1))
        
        cov_mat_com=np.zeros((n,n))
       
        #print(t_mean.shape)
        
        class_cnt=[]
        for i in range(k):
            t_mean[i,:,0]=mean_list[i]
            class_cnt.append(0)
       
        for i in range(X.shape[0]):
            for j in range(len(class_names)):
                if(y[i] == class_names[j]):
                    t1=np.array([X[i,:]])
                    tmp=(t1.T-t_mean[j]).dot((t1.T-t_mean[j]).T)
                    #tmp=(X[i,:].T-t_mean[j]).dot((X[i,:].T-t_mean[j]).T)
                    cov_mat_com=cov_mat_com+tmp
                    #tmp=(X[i,:]-t_mean[j]).dot((X[i,:]-t_mean[j]).T)
                    break
        cov_mat_com=cov_mat_com/(X.shape[0])
        return cov_mat_com
                
    def find_distance(self,X,x,dist_type,y,mean_list,class_names):
        N=X.shape[0]
        n=X.shape[1]
        dist_list=[]
        #sigma=self.find_co_variance(X,y,mean_list,class_names)
        #sigma_inv=np.linalg.inv(sigma)
        for i in range(N):
            if(dist_type=='Euclidean'):
                val=abs(((X[i]-x).T).dot((X[i]-x)))
                dist=math.sqrt(val)
            #elif(dist_type=="Mahalonabis"):
                
                #dist=distance.mahalanobis(X[i],x,sigma_inv)
            elif(dist_type=='Manhattan'):
                dist=0
                for j in range(n):
                    dist +=abs(X[i,j]-x[j])
            dist_list.append(dist)
                
        return dist_list
    def get_mean(self,X):
        return np.mean(X,axis=0)
    
    def predict(self,y,dist_list,class_names,k):
        dist_list,y_tmp = zip(*sorted(zip(dist_list, y)))
        
        neighbour_cnt=[]
        for i in range(len(class_names)):
            neighbour_cnt.append(0)
        
        for i in range(1,k):
            for j in range(len(class_names)):
                if(y_tmp[i]==class_names[j]):
                    neighbour_cnt[j] +=1
                    break
        pred_class=np.argmax(neighbour_cnt)
        return class_names[pred_class]
                    
#         return preds
    
    def write_prediction(self,filename,preds):
        param_file=open(filename,"w")
        for i in range(0,len(preds)):
            param_file.write(str(preds[i]))
            param_file.write("\n")
        param_file.close()
    
    def get_classifications(self,y,preds):
        preds_val=[]
        correct=0
        incorrect=0
        tp=0
        tn=0
        fp=0
        fn=0
        
        for i in range(len(preds)):
            if(preds[i] ==y[i]):
                if(y[i]==1):
                    tp +=1
                else:
                    tn+=1
            else:
                if(y[i]==1):
                    fn +=1
                else:
                    fp+=1
        return tp,tn,fp,fn   
    def get_perf_params(self,tp,tn,fp,fn):
        sensitivity=0
        specificity=0
        precision=0
        f_measure=0
        if(tp+fn !=0):
            sensitivity=tp/(tp+fn)
        if(tn+fp !=0):
            specificity=tn/(tn+fp)
        if(tp+fp !=0):
            precision=tp/(tp+fp)
        if(precision + sensitivity !=0):
            f_measure=(2*precision*sensitivity)/(precision+sensitivity)
        accuracy=(tp+tn)/(tp+tn+fp+fn)
        
        return accuracy,sensitivity,specificity,precision,f_measure
    
    def plot_ROC(self,neg_specificity_list,sensitivity_list):
        plt.title('Receiver Operating Characterestic Curve');
        plt.xlabel('1-Specificity')
        plt.ylabel('Sensitivity')
        plt.xlim(-0.1,1.05)
        plt.ylim(-0.1,1.05)
        plt.plot(neg_specificity_list,sensitivity_list,label='ROC',color='r')
        plt.show()
        
    
    def cross_validation(self,X,y,class_names,mean_list,k_min,k_max,step_size):
        
        N=X.shape[0]
        n=X.shape[1]
        #print(N,n)
        accuracy_list=[]
        sensitivity_list=[]
        precision_list=[]
        neg_specificity_list=[]
        f_measure_list=[]
        
        k_arr=[]
        for k_val in range(k_min,k_max,step_size):  
            print("k=",k_val)
            preds=[]
            k_arr.append(k_val)
            k_int=int(k_val)
            for i in range(N):
                dist_list=knn.find_distance(X,X[i],"Euclidean",y,mean_list,class_names)
                #dist_list=knn.find_distance(X,X[i],"Mahalonabis",y,mean_list,class_names)
                #dist_list=knn.find_distance(X,X[i],"Manhattan",y,mean_list,class_names)                
                preds.append(knn.predict(y,dist_list,class_names,k_int))
                          
            
            tp,tn,fp,fn=self.get_classifications(y,preds)
            accuracy,sensitivity,specificity,precision,f_measure=self.get_perf_params(tp,tn,fp,fn)
            accuracy_list.append(accuracy)
            f_measure_list.append(f_measure)
            sensitivity_list.append(sensitivity)
            neg_specificity_list.append(1-specificity)
            
        index=f_measure_list.index(max(f_measure_list))
        return k_arr[index],max(f_measure_list),neg_specificity_list,sensitivity_list
      
    
    
```
