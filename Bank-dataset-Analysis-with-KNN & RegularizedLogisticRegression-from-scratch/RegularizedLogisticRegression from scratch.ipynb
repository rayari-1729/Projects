```python
class RegularizedLogisticRegressionModel:
    def __init__(self, lr=0.01, num_iter=100000, fit_intercept=True,lagrange_mul=1):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.lagrange_mul=lagrange_mul
    
    def add_intercept(self, X):
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    def sigmoid(self, z):        
        return 1 / (1 + np.exp(-z))
        
    def grad(self, X, y,w,w0,lagrange_mul):
        N=X.shape[0]
        #print(w0)
        tmp=0
        tmp2=0
        
        for i in range(N):
            z=X[i].dot(w)+w0
            f_x_i=self.sigmoid(z)
            tmp +=f_x_i-y[i]
            #print(tmp)
            tmp3=np.array([X[i]])
            #print(tmp3.shape)
            tmp2 +=(tmp3.T).dot((f_x_i-y[i]))
        
        tmp2=np.array([tmp2])
        tmp2=tmp2.T
        
        w0=w0-(self.lr/N)*(tmp)
        w=w*(1-(self.lr*lagrange_mul)/N) - (self.lr/N)*(tmp2)
        return w,w0 
    
    def predict_prob(self, X,w):
        if self.fit_intercept:
            X = self.add_intercept(X)
        #w=w/X.shape[0]
        z=np.dot(X, w)
        
        result=self.sigmoid(z)
        #print(result)
        return result
    
    def predict(self, X, threshold,w):
        return self.predict_prob(X,w) >= threshold
    
    def loss(self, f_x, y,w,lagrange_mul):
        n=w.shape[0]
        N=f_x.shape[0]
        
        tmp=0
        tmp2=0
        for i in range(1,n):
            tmp +=w[i]**2
        for i in range(N):
            tmp2 +=y[i]*np.log(f_x[i]) + (1-y[i])*np.log(1-f_x[i])
        j_w= -tmp2+lagrange_mul*tmp       
        return j_w/N
    
    def write_prediction(self,filename,preds):
        param_file=open(filename,"w")
        for i in range(0,len(preds)):
            if(preds[i]==True):
                param_file.write(str(1))
            else:
                param_file.write(str(0))
            param_file.write("\n")
        param_file.close()
    
    def get_classifications(self,y,preds):
        preds_val=[]
        tp=0
        tn=0
        fp=0
        fn=0
        for i in range(len(preds)):
            if(preds[i] == True):
                preds_val.append(1)
            else:
                preds_val.append(0)
            if(preds_val[i] ==y[i]):
                if(preds_val[i]==1):
                    tp=tp+1
                else:
                    tn=tn+1
            else:
                if(preds_val[i]==1):
                    fp=fp+1
                else:
                    fn=fn+1  
        
        return tp,tn,fp,fn   
    def get_perf_params(self,tp,tn,fp,fn):
        sensitivity=0
        specificity=0
        precision=0
        f_measure=0
        if(tp+fn !=0):
            sensitivity=tp/(tp+fn)
        if(tn+fp !=0):
            specificity=tn/(tn+fp)
        if(tp+fp !=0):
            precision=tp/(tp+fp)
        if(precision + sensitivity !=0):
            f_measure=(2*precision*sensitivity)/(precision+sensitivity)
        accuracy=(tp+tn)/(tp+tn+fp+fn)
        
        return accuracy,sensitivity,specificity,precision,f_measure
```
