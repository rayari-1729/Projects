## Find and display the count of medals (Bronze, Silver, Gold) won in each city, and [resent the result ina  atable format 

```python
data.head(1)
```

```python
data.groupby(['City', 'Medal']).size().unstack() # convert the multilabel fata to pandas format 
```

```python
medal_count = data.groupby(['City', 'Medal']).size().unstack()
medal_count['Total'] = medal_count[['Bronze', 'Gold', 'Silver']].sum(axis=1)
```

```python
medal_count.sort_values(by= 'Total', ascending = False) ##<======================
```
